## Python

```python
image = cv2.imread('image1.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)

harris_corners = cv2.cornerHarris(gray,21,21,0.01)
corners = np.where(harris_corners > 0.25 * harris_corners.max())

plt.imshow(image)
plt.title('corners')
plt.scatter(x=corners[1], y=corners[0], c='r')
plt.show()
```

```python

criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001)

objp = np.zeros((6*7,3), np.float32)
objp[:,:2] = np.mgrid[0:7,0:6].T.reshape(-1,2)

objpoints = []
imgpoints = []
images = ['image1.jpg', 'image2.jpg']

image_corners = []


for fname in images:
    img = cv2.imread(fname)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    ret, corners = cv2.findChessboardCorners(gray, (6,8), None)
    if ret == True:
        image_corners.append(corners)
        objpoints.append(objp)
        corners2 = cv2.cornerSubPix(gray,corners, (11,11), (-1,-1), criteria)
        imgpoints.append(corners)
        cv2.drawChessboardCorners(img, (6,8), corners2, ret)
        plt.imshow(img)
        plt.title('img')
        plt.show()
```

```python
p = image_corners[1]

square_x_mm = 280
square_y_mm = 280

image_point = []
world_point = []

for i in range(8):
    for j in range(6):
        image_point.append(p[i*6 + j][0])
        world_point.append([j*square_x_mm, i*square_y_mm, 0])

world_point = np.reshape(world_point, (48,3))
image_point = np.reshape(image_point, (48,2))
```

```python
def build_line(image_points, world_points, i):
    return [
        image_points[i][0]*world_points[i][0],
        image_points[i][0]*world_points[i][1],
        image_points[i][0]*world_points[i][2],
        image_points[i][0],
        -1*image_points[i][1]*world_points[i][0],
        -1*image_points[i][1]*world_points[i][1],
        -1*image_points[i][1]*world_points[i][2],
        -1*image_points[i][1],
    ]
```

```python
A = []

for i in range(48):
    A.append(build_line(image_point, world_point, i))

A = np.reshape(A,(48,8))

SVD = np.linalg.svd(A, full_matrices=True)

print(SVD[1])
print("=====")
print(SVD[2])
```
